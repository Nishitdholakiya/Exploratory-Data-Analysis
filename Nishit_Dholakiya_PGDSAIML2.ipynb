## Q1: Data Cleaning:
# i. Count and handle Missing values.

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv("/Users/dholakiyan/Desktop/HeroAssign/superstore.csv")

# Check missing values
missing_values = df.isnull().sum()
print(missing_values)

# Percentage of missing values
missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent)

# Handling missing values
# Example: Fill categorical with mode, numerical with median

for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].median(), inplace=True)
```

Conclusion:
The dataset contains minimal missing values, with only Shipping.Cost and State having negligible missing entries (<1%). These were handled using median and mode imputation respectively. Overall, the dataset is clean and requires minimal preprocessing.

# ii. Deal with Duplicate Values

```python
duplicates = df.duplicated().sum()
print("Duplicate rows:", duplicates)
```

Conclusion:
